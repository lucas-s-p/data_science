## Gerando o novo Dataset

```python
data = pd.DataFrame(iris.data[:, :1], columns =['sepal_length'])
df1 = data[0:50]
df1
```

## Definição da função de Bootstrap

Dependendo da forma como você pretenda resolver o problema, você pode ter interesse em alterar a função bootstrap disponibilizada abaixo. Fique à vontade para gerar variações dessa função se assim desejar.

```python
def bootstrap(data, func, alpha=0.95, n_rep=1000):

  theta_hat_error_list = list()
  theta = func(data)

  # Resampling
  for i in range(0,n_rep):
    x = resample(data, replace=True)
    theta_hat_error = func(x) - theta
    theta_hat_error_list.append(theta_hat_error)

  # Percentiles
  lower_p = ((1.0-alpha)/2.0) * 100
  upper_p = (alpha+((1.0-alpha)/2.0)) * 100

  # Lower and Upper Bounds
  lower_value = np.percentile(theta_hat_error_list, lower_p)
  upper_value = np.percentile(theta_hat_error_list, upper_p)

  lower_limit = theta+lower_value
  upper_limit = theta+upper_value

  return theta, lower_limit, upper_limit
```

## Análise das diferenças entre os dois tipos de flor escolhidos.

```python
df_diff = pd.DataFrame()
df_diff = df1.apply(lambda row: df2['sepal_length'] - df1['sepal_length'])
df_diff
```

```python
alpha = 0.95
theta_bs, lower_limit, upper_limit = bootstrap(df_diff, np.mean)

print('Theta : %.3f' % theta_bs)
print('Confidence Interval (%.f%%): %.3f to %.3f' % (alpha*100, lower_limit, upper_limit))
```

Resultado:

Podemos afirmar que com 95% de confiança, em média, as flores do tipo 2 (df2)  possuem um tamanho de sépala maior que as flores do tipo 1 (df1) em 0.930cm, com intervalo de confiança de 0.748cm a 1.118cm

```python
# Variação do bootstrap, outra forma de calcular o intervalo de confiança.

def bootstrap_mod(data1, data2, func, alpha=0.95, n_rep=1000):

  theta_hat_error_list = list()
  theta = func(data2-data1)

  # Resampling
  for i in range(0,n_rep):
    x1 = resample(data1, replace=True)
    x2 = resample(data2, replace=True)
    theta_hat_error = func(x2 - x1) - theta
    theta_hat_error_list.append(theta_hat_error)

  # Percentiles
  lower_p = ((1.0-alpha)/2.0) * 100
  upper_p = (alpha+((1.0-alpha)/2.0)) * 100

  # Lower and Upper Bounds
  lower_value = np.percentile(theta_hat_error_list, lower_p)
  upper_value = np.percentile(theta_hat_error_list, upper_p)

  lower_limit = theta+lower_value
  upper_limit = theta+upper_value

  return theta, lower_limit, upper_limit
```

```python
alpha = 0.99
theta_bs, lower_limit, upper_limit = bootstrap_mod(df1, df2, np.mean)

print('Theta : %.3f' % theta_bs)
print('Confidence Interval (%.f%%): %.3f to %.3f' % (alpha*100, lower_limit, upper_limit))
```



Resultado:

Podemos afirmar que com 99% de confiança, em média, as flores do tipo 2 (df2) possuem um tamanho de sépala maior que as flores do tipo 1 (df1) em 0.930cm, com intervalo de confiança de 0.649cm a 1.215cm
