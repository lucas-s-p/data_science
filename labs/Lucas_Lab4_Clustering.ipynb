## Primeiro Agrupamento (Cálculo + Plot)

### Definindo a entrada do algoritmo (X)

```python
X = df[['Renda', 'Pontuacao Consumo']].values
```

### Cálculo

```python
# código

algorithm = KMeans(n_clusters = 5)
algorithm.fit(X)
labels = algorithm.labels_
centroids = algorithm.cluster_centers_
```

### Plot

```python
y_kmeans = algorithm.predict(X)
plt.scatter(X[:, 0], X[:, 1], c=y_kmeans, s=50, cmap='plasma')

centers = algorithm.cluster_centers_
plt.scatter(centers[:, 0], centers[:, 1], c='black', s=200, alpha=0.5);
```

## Segundo Agrupamento (Cálculo + Plot)

```python
X1 = df[['Idade', 'Pontuacao Consumo']].values
```

```python
# código

algorithm1 = KMeans(n_clusters = 4, init='k-means++')
algorithm1.fit(X1)
labels = algorithm1.labels_
centroids = algorithm1.cluster_centers_
```
